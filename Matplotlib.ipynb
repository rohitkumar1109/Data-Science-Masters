```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(6,3))
plt.scatter(x,y, c='#FFB833', alpha=0.9, marker='x')
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
plt.title("Scatter Plot of x vs y")
plt.grid()
```

```python
x = np.linspace(1,10, 100)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.plot(x,y, c='r')
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x,y)
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y, '--r')
plt.xlabel('This is my x axis')
plt.ylabel('This is my y axis')
plt.title('Sin graph of x vs y')
plt.grid()
```

```python
x = ['a','b','c','d','e']
y = np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize= (8,3))
plt.bar(x,y, color='b',edgecolor='r')
plt.xlabel('This is my x axis')
plt.ylabel('This is my y axis')
plt.title('Bar Graph')
plt.grid()
```

```python
plt.figure(figsize= (8,3))
plt.barh(x,y, color='b',edgecolor='r')
plt.xlabel('This represents numerical data')
plt.ylabel('This represents categorical data')
plt.title('Bar Graph')
plt.grid()
```

```python
data = [1,2,3,3,4,4,5,5,6,6,5,6,7,7,7,8,8,8,8,8]
```

```python
plt.figure(figsize=(6,3))
plt.hist(data, color='b')
plt.xlabel('Represents unique values')
plt.ylabel('Represents frequency of values')
plt.title('Histogram plot example')
plt.grid()
plt.savefig("test.png")
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
fig = plt.figure()
axis = fig.add_subplot(projection = '3d')
axis.scatter(x,y,z, c='b', alpha= 0.8)
plt.xlabel('This is my x-axis')
plt.ylabel('This is my y-axis')
plt.title('3D Graph Example')
plt.show()
```
